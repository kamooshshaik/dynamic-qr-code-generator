```python
!pip install qrcode[pil] pillow requests
# ðŸ“š Imports
import qrcode
from PIL import Image, ImageDraw
from google.colab import files

# âœ… Step 1: Setup URL
url = "https://www.iitk.ac.in/ce/"

# âœ… Step 2: Generate QR code
qr = qrcode.QRCode(
    version=4,
    error_correction=qrcode.constants.ERROR_CORRECT_H,  # High correction for logo
    box_size=10,
    border=4,
)
qr.add_data(url)
qr.make(fit=True)
qr_img = qr.make_image(fill_color="black", back_color="white").convert("RGB")

# âœ… Step 3: Upload logo image (JPEG or PNG)
uploaded = files.upload()  # Upload your logo or any small image
logo_path = list(uploaded.keys())[0]
logo = Image.open(logo_path).convert("RGB")  # âœ… Force RGB, no transparency required

# âœ… Step 4: Resize logo
qr_width, qr_height = qr_img.size
logo_size = int(qr_width * 0.2)
logo = logo.resize((logo_size, logo_size))

# Use smaller logo
logo_size = int(qr_width * 0.15)

# Add padding around logo
padding = 10
padded_size = logo_size + 2 * padding
logo_bg = Image.new("RGB", (padded_size, padded_size), "white")
logo = logo.resize((logo_size, logo_size))
logo_bg.paste(logo, (padding, padding))

# Paste centered logo
pos = ((qr_width - padded_size) // 2, (qr_height - padded_size) // 2)
qr_img.paste(logo_bg, pos)


# âœ… Step 6: Save final image
output_filename = "dynamic_qr_with_logo.jpeg"
qr_img.save(output_filename, "JPEG")

# âœ… Step 7: Display + download
qr_img.show()
files.download(output_filename)
```
